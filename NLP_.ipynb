```python
import pandas as pd
df=pd.read_csv('food_network.csv')
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
docs=df['text']
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_df=0.8, min_df=0.02)
tfidf = tfidf_vectorizer.fit_transform(docs)

tfidf_df = pd.DataFrame(tfidf.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
tfidf_df
```

```python
# Apply NMF
from sklearn.decomposition import NMF

nmf_model = NMF(n_components=20, random_state=0)
#OpenAI
nmf_model.fit(tfidf)
W= nmf_model.fit_transform(tfidf)  # Document-topic matrix

# Display topics
feature_names = tfidf_vectorizer.get_feature_names_out()
topic_names=[]
for topic_index in range(len(nmf_model.components_)):
    topic = nmf_model.components_[topic_index]
    sorted_indices = topic.argsort()
    top_indices = sorted_indices[-4:]
    top_indices = top_indices[::-1]

    top_words = []
    for i in top_indices:
        top_words.append(feature_names[i])

    top_words_string = " ".join(top_words)

    print("Topic #{}:".format(topic_index))
    print(top_words_string)
    topic_names.append(top_words_string)

topic_df = pd.DataFrame(W, columns=topic_names)
topic_df
```

```python
# Code for estimating the linear regression model with the outcome being $log(Likes)$
import pingouin as pg
import numpy as np

# Combine X and y into a single dataframe
# OpenAI
df_model = topic_df.copy()
df_model['like'] =np.log1p(df['public_metrics.like_count']+1)

# Run linear regression
result = pg.linear_regression(df_model.drop(columns='like'), df_model['like'])

# Display R-squared (only need the first row since it's repeated)
r2_1 = round(result['r2'].iloc[0], 3)
print(f"\nR-squared: {r2_1}")

# Round coef and pval to 3 decimal places
result[['names', 'coef', 'pval']] = result[['names', 'coef', 'pval']].round(3)

# Display the rounded result
result[['names', 'coef', 'pval']]
```
